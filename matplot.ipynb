# Matplotlib

Matplotlib is a powerful plotting library for python that enables the creation of static,animated and interactive visualizations.
It is widely used for data visualization in data science and analytics.

```python
import matplotlib.pyplot as plt
import pandas as pd

x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Create a line plot
plt.plot(x,y)
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Basic Line Plot')
plt.show
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]

# Create a customized line plot
plt.plot(x,y,color='red')
plt.show()

# Plot with linestyle as "--(double dash)"
plt.plot(x,y,color='red',linestyle='--')
plt.show()

plt.plot(x,y,color='blue',linewidth=3)
plt.show()

# To enable grid in the plot
plt.plot(x,y)
plt.grid(True)
plt.show()
```

```python
# Multiple PLots 
# Sample Data

x = [1,2,3,4,5]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]

plt.figure(figsize=(9,5))

# Plot with 1 row and 2 cols
# Plot in the first row first col
plt.subplot(2,2,1)
plt.plot(x,y1,color='green')
plt.title("Plot 1")

# Plot in the first row second col
plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.title('Plot 2')

# Plot in the second row first col
plt.subplot(2,2,3)
plt.plot(y1,x,color='yellow')
plt.title('Plot 3')

# Plot in the second row second col
plt.subplot(2,2,4)
plt.plot(y2,x,color='orange')
plt.title('Plot 4')
```
